```python
import matplotlib.pyplot as plt #mengimport library matplotlib
%matplotlib inline

from skimage import data #mengimport library skimage
from skimage import io #mengimport library skimage
from skimage.util import invert #mengimport library skimage

import numpy as np #mengimport library numpy
```

```python
foto1 = io.imread ("foto/lucu.jpeg") #memanggil gambar pada folder
foto2 = io.imread ("foto/kean.jpeg") #memanggil gambar pada folder

foto1_Cropped = foto1.copy() 
foto1_Cropped = foto1_Cropped [0:246,64:310] #mengatur ukuran foto1 untuk dicrop

foto2_Cropped = foto2.copy()
foto2_Cropped = foto2_Cropped [64:246,128:310] #mengatur ukuran foto2 untuk dicrop

print('R25 Ori Shape : ',foto1.shape) #mencetak gambar asli foto1
print('R25 Crop Shape : ',foto1_Cropped.shape) #mencetak gambar crop foto1

print('Mobil Ori Shape : ',foto2.shape) #mencetak gambar asli foto1
print('Mobil Crop Shape : ',foto2_Cropped.shape) #mencetak gambar crop foto1

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(foto1) #memanggil foto1
ax[0].set_title("Citra Input 1") #menamai output gambar

ax[1].imshow(foto2, cmap='gray') #memanggil foto2
ax[1].set_title('Citra Input 2') #menamai output gambar

ax[2].imshow(foto1_Cropped) #memanggil foto1 yang sudah dicrop
ax[2].set_title("Citra Output 1") #menamai output gambar

ax[3].imshow(foto2_Cropped, cmap='gray') #memanggil foto2 yang sudah dicrop
ax[3].set_title('Citra Output 2') #menamai output gambar
```

```python
inv = invert(foto1_Cropped) #menginvert gambar foto1
print('Shape Input : ', foto1_Cropped.shape) #mengeksekusi gambar foto1 crop
print('Shape Output : ',inv.shape)

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(foto1_Cropped) #memanggil foto1 yang sudah dicrop
ax[0].set_title("Citra Input") #menamai output gambar

ax[1].hist(foto1_Cropped.ravel(), bins=256) #menampilkan histogram foto1 yang sudah dicrop
ax[1].set_title('Histogram Input') #menamai output gambar

ax[2].imshow(inv) #menampilkan gambar yang sudah diinvert (menjadi citra negative)
ax[2].set_title('Citra Output (Inverted Image)') #menammai gambar

ax[3].hist(inv.ravel(), bins=256) #menampilkan histogram negative foto1 yang sudah dicrop
ax[3].set_title('Histogram Output')
```

```python
copyFoto = foto2_Cropped.copy().astype(float)

shape = copyFoto.shape
output1 = np.empty(shape)

for baris in range(0, shape[0]):
    for kolom in range(0, shape[1]):
        a1 = baris
        b1 = kolom
        output1[a1, b1] = copyFoto[baris, kolom] /100
        
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(foto2_Cropped, cmap='gray') #memanggil gambar foto2 yang sudah dicrop
ax[0].set_title("Citra Input") #menamai gambar

ax[1].hist(foto1_Cropped.ravel(), bins=256) #menampilkan gambar histogram foto1 yang sudah dicrop
ax[1].set_title('Histogram Input') #menamai gambar

ax[2].imshow(output1, cmap='gray') #menampilkan gambar brightness
ax[2].set_title('Citra Output (Brightnes)') #menamai gambar

ax[3].hist(output1.ravel(), bins=256) #menampilkan histogram 
ax[3].set_title('Histogram Input') #menamai gambar
```
